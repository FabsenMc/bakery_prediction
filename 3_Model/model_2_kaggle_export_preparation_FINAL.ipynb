```python
# import csv-files of the kaggel competition and merge them with the data from the data preparation

import pandas as pd
import numpy as np

df_1 = pd.read_csv('/workspaces/bakery_prediction/0_DataPreparation/test_dataf_knn.csv')
df_1['Datum'] = pd.to_datetime(df_1['Datum'])
df_1 = df_1.drop('Warengruppe', axis=1)
df_2 = pd.read_csv('/workspaces/bakery_prediction/3_Model/sample_submission.csv')

# Normalization of "Windgeschwindigkeit", 35 = max value
df_1['Windgeschwindigkeit_N'] = (df_1['Windgeschwindigkeit'] - df_1['Windgeschwindigkeit'].min()) / (35 - df_1['Windgeschwindigkeit'].min())

# Normalization of "Verbraucherpreisindex"
df_1['Verbraucherpreisindex_N'] = (df_1['Verbraucherpreisindex'] - 93.1) / (100.3 - 93.1)

# Create the columns for Temperatur_Kategorie
dummies = pd.get_dummies(df_1['Temperatur_Kategorie'], prefix='Temperatur_Kategorie')
df_1 = pd.concat([df_1, dummies], axis=1)
df_1= df_1.replace({True: 1, False: 0})

# Create the columns for Jahreszeit_FSHW
dummies = pd.get_dummies(df_1['Jahreszeit_FSHW'], prefix='Jahreszeit_FSHW')
df_1 = pd.concat([df_1, dummies], axis=1)
df_1 = df_1.replace({True: 1, False: 0})

# Create the columns for Wochentag_MDMDFSS
df_1['Wochentag_MDMDFSS'] = df_1['Wochentag_MDMDFSS'].astype('int64')
dummies = pd.get_dummies(df_1['Wochentag_MDMDFSS'], prefix='Wochentag_MDMDFSS')
df_1 = pd.concat([df_1, dummies], axis=1)
df_1 = df_1.replace({True: 1, False: 0})

# Make the correct date out of the id
df_2['id'] = df_2['id'].astype(str)
df_2['Datum'] = pd.to_datetime(df_2['id'].str[:6], format='%y%m%d')
df_2['Datum'] = pd.to_datetime(df_2['Datum'])

# Create the columns for the Warengruppen
df_2['Warengruppe'] = df_2['id'].str[-1]
dummies = pd.get_dummies(df_2['Warengruppe'], prefix='Warengruppe')
df_2 = pd.concat([df_2, dummies], axis=1)
df_2 = df_2.replace({True: 1, False: 0})

# Join the csv-files
df = pd.merge(df_2, df_1, how='left', on='Datum')

# Add 'Saison_Warengruppe' for 'Warengruppe' 6
df['Saison_Warengruppe'] = np.where(df['Warengruppe'] == 6, 1, 0)

df.fillna(0, inplace=True)

df['Jahreszeit_FSHW_2'] = df['Jahreszeit_FSHW_2'].astype('int64')
df['Jahreszeit_FSHW_3'] = df['Jahreszeit_FSHW_3'].astype('int64')
df['Jahreszeit_FSHW_4'] = df['Jahreszeit_FSHW_4'].astype('int64')

df['Wochentag_MDMDFSS_2'] = df['Wochentag_MDMDFSS_2'].astype('int64')
df['Wochentag_MDMDFSS_3'] = df['Wochentag_MDMDFSS_3'].astype('int64')
df['Wochentag_MDMDFSS_4'] = df['Wochentag_MDMDFSS_4'].astype('int64')
df['Wochentag_MDMDFSS_5'] = df['Wochentag_MDMDFSS_5'].astype('int64')
df['Wochentag_MDMDFSS_6'] = df['Wochentag_MDMDFSS_6'].astype('int64')

df['Temperatur_Kategorie_2'] = df['Temperatur_Kategorie_2'].astype('int64')
df['Temperatur_Kategorie_3'] = df['Temperatur_Kategorie_3'].astype('int64')

df['FerienSH'] = df['FerienSH'].astype('int64')

df['Schnee'] = df['Schnee'].astype('int64')

print(df.head())

# Export csv file
df_export = df[['id', 'Datum', 'Warengruppe_2', 'Warengruppe_3', 'Warengruppe_4', 'Warengruppe_5', 'Warengruppe_6', 'Temperatur_Kategorie_2', 'Temperatur_Kategorie_3', 'Jahreszeit_FSHW_2', 'Jahreszeit_FSHW_3', 'Jahreszeit_FSHW_4', 'Wochentag_MDMDFSS_2', 'Wochentag_MDMDFSS_3', 'Wochentag_MDMDFSS_4', 'Wochentag_MDMDFSS_5', 'Wochentag_MDMDFSS_6', 'FerienSH', 'Feiertag', 'Saison_Warengruppe', 'Verbraucherpreisindex_N', 'Schnee']]
                
df_export.to_csv('kaggle_sample_nn_final.csv', index=False)
```
